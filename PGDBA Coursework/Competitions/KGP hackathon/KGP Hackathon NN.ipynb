```python
#importing relevant libraries
import numpy as np
import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing as prep
from sklearn.metrics import r2_score
from tensorflow import keras
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
from tensorflow.keras.layers import Dense
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
#importing train and holdout datasets
data = pd.read_csv(r"C:\Users\HP\Desktop\PGDBA IIM-ISI-IIT\Competitions\Kharagpur Data Science Hackathon\Data\train.csv")
holdout = pd.read_csv(r"C:\Users\HP\Desktop\PGDBA IIM-ISI-IIT\Competitions\Kharagpur Data Science Hackathon\Data\test.csv")
```

```python
#feature engineering

# month-year
# data['year_label'] = pd.to_datetime(data['placement date'], format = "%d-%m-%Y").dt.year-2017
# data['month_label'] = pd.to_datetime(data['placement date'], format = "%d-%m-%Y").dt.month
# data['month_num'] = data['year_label']*data['month_label']

# holdout['year_label'] = pd.to_datetime(holdout['placement date'], format = "%Y-%m-%d").dt.year-2017
# holdout['month_label'] = pd.to_datetime(holdout['placement date'], format = "%Y-%m-%d").dt.month
# holdout['month_num'] = holdout['year_label']*holdout['month_label']

# data['placement date'] = data['year_label']*data['month_label']

# data['price'] = np.log(data['price'])
# holdout['price'] = np.log(holdout['price'])

data['cost'] = data['price']*(1-data['discount'])
holdout['cost'] = holdout['price']*(1-holdout['discount'])

data['total_cost'] = data['cost'] * data['no of items']
holdout['total_cost'] = holdout['cost'] * data['no of items']

# data['total_cost'] = np.sqrt(data['total_cost'])
# holdout['total_cost'] = np.sqrt(holdout['total_cost'])

data['year_label'] = (pd.to_datetime(data['placement date'], format = "%d-%m-%Y").dt.year-2017)/2021
data['month_label'] = (pd.to_datetime(data['placement date'], format = "%d-%m-%Y").dt.month)/12

holdout['year_label'] = (pd.to_datetime(holdout['placement date'], format = "%Y-%m-%d").dt.year-2017)/2021
holdout['month_label'] = (pd.to_datetime(holdout['placement date'], format = "%Y-%m-%d").dt.month)/12

bins = [-1.9,1,1.7,2.1,2.4,2.7,2.9,3.1,3.4,3.6,3.8,4.1,4.4,4.6,4.9,5.2,5.5,5.8,6.2,6.7,9.8]
data['cost_labels'] = pd.cut(np.log(data['cost']), bins = bins)
holdout['cost_labels'] = pd.cut(np.log(holdout['cost']), bins = bins)

#-------------------------------------------------------------------------------#
#defining categorical and numerical dataframes
data_cat = data[['departure city','location','class','segment','sub-class','departure state','cost_labels']].astype(str)
data_num = data[['discount','no of items', 'price','cost','total_cost']]

holdout_cat = holdout[['departure city','location','class','segment','sub-class','departure state','cost_labels']].astype(str)
holdout_num = holdout[['discount','no of items', 'price','cost','total_cost']]

#preparing ohe and le-
enc = prep.OneHotEncoder()
le = prep.LabelEncoder()

#one hot encoding the categorical data
data_cat = data_cat.apply(le.fit_transform)
holdout_cat = holdout_cat.apply(le.fit_transform)

enc.fit(data_cat)

data_cat    = pd.DataFrame.sparse.from_spmatrix(enc.transform(data_cat.apply(le.fit_transform)))
# holdout_cat = pd.DataFrame.sparse.from_spmatrix(enc.transform(holdout_cat.apply(le.fit_transform)))

#Preparing X and H
X = data_num.join(data_cat)
H = holdout_num.join(holdout_cat)
y = data['profit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
def baseline_model():
    # create model
    model = Sequential()
    model.add(Dense(128, input_dim=X_train.shape[1], kernel_initializer='normal', activation='relu'))
    model.add(Dense(64, kernel_initializer='normal', activation='relu'))
    model.add(Dense(16, kernel_initializer='normal', activation='relu'))
    model.add(Dense(8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, kernel_initializer='normal', activation = 'linear'))
    # Compile model
    
    
    opt = keras.optimizers.Adam(learning_rate=0.0002)
    model.compile(loss='mae', optimizer=opt, metrics='mae')
    return model
```

```python
# evaluate model
model = KerasRegressor(build_fn=baseline_model)
# kfold = KFold(n_splits=10)
# results = cross_val_score(estimator, X, y)
# print("Baseline: %.2f (%.2f) MSE" % (results.mean(), results.std()))

model.fit(np.array(X_train), y_train, validation_data = (X_test, y_test), \
                          epochs=300,verbose=1)
```

# R2 - NN

```python
prediction = model.predict(X_test)
r2_score(y_test, prediction)
```
