```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import nltk

from nltk.tokenize import word_tokenize
from nltk.text import Text

import os
```

```python
f1 = open('FileNegative.txt', 'r', encoding="utf8")
f2 = open('FilePositive.txt', 'r', encoding="utf8")
f3 = open('ajantapharma_17-18.txt', 'r', encoding="utf8")
f4 = open('ajantapharma_18-19.txt', 'r', encoding="utf8")
f5 = open('ajantapharma_19-20.txt', 'r', encoding="utf8")
f6 = open('Granules_17-18.txt', 'r', encoding="utf8")
f7 = open('Granules_18-19.txt', 'r', encoding="utf8")
f8 = open('Granules_19-20.txt', 'r', encoding="utf8")

FileNegative = f1.read().lower()
FilePositive = f2.read().lower()
ajantapharma_17_18 = f3.read().lower()
ajantapharma_18_19 = f4.read().lower()
ajantapharma_19_20 = f5.read().lower()
granules_17_18 = f6.read().lower()
granules_18_19 = f7.read().lower()
granules_19_20 = f8.read().lower()
```

# Ajanta Pharma

```python
def findCount(tokens,report_text):
    total_count=0
    for token in tokens:
        count = report_text.vocab().get(token)
        if(count!= None ):
#             print('token:',token, ' count:', count)
            total_count = total_count+count
    print('Total count:',total_count)
    return total_count
```

```python
ajantapharma_17_18_neg = findCount(tokens_negative,text_ajantapharma_17_18)
```

## Ajanta pharma

```python
# Without normalisation

ajantapharma_17_18_neg = findCount(tokens_negative,text_ajantapharma_17_18)
ajantapharma_17_18_pos = findCount(tokens_positive,text_ajantapharma_17_18)

ajantapharma_18_19_neg = findCount(tokens_negative,text_ajantapharma_18_19)
ajantapharma_18_19_pos = findCount(tokens_positive,text_ajantapharma_18_19)

ajantapharma_19_20_neg = findCount(tokens_negative,text_ajantapharma_19_20)
ajantapharma_19_20_pos = findCount(tokens_positive,text_ajantapharma_19_20)
```

```python
# With normalisation

ajantapharma_17_18_neg = findCount(tokens_negative,text_ajantapharma_17_18)/len(tokens_ajantapharma_17_18)
ajantapharma_17_18_pos = findCount(tokens_positive,text_ajantapharma_17_18)/len(tokens_ajantapharma_17_18)

ajantapharma_18_19_neg = findCount(tokens_negative,text_ajantapharma_18_19)/len(tokens_ajantapharma_18_19)
ajantapharma_18_19_pos = findCount(tokens_positive,text_ajantapharma_18_19)/len(tokens_ajantapharma_18_19)

ajantapharma_19_20_neg = findCount(tokens_negative,text_ajantapharma_19_20)/len(tokens_ajantapharma_19_20)
ajantapharma_19_20_pos = findCount(tokens_positive,text_ajantapharma_19_20)/len(tokens_ajantapharma_19_20)
```

# Granules

```python
tokens_granules_17_18= word_tokenize(granules_17_18)
text_granules_17_18 = Text(tokens_granules_17_18)

tokens_granules_18_19= word_tokenize(granules_18_19)
text_granules_18_19 = Text(tokens_granules_18_19)

tokens_granules_19_20= word_tokenize(granules_18_19)
text_granules_19_20 = Text(tokens_granules_19_20)
```

```python
# Without Normalisation
granules_17_18_neg = findCount(tokens_negative,text_granules_17_18)
granules_17_18_pos = findCount(tokens_positive,text_granules_17_18)

granules_18_19_neg = findCount(tokens_negative,text_granules_18_19)
granules_18_19_pos = findCount(tokens_positive,text_granules_18_19)

granules_19_20_neg = findCount(tokens_negative,text_granules_19_20)
granules_19_20_pos = findCount(tokens_positive,text_granules_19_20)
```

```python
# With normalisation
granules_17_18_neg = findCount(tokens_negative,text_granules_17_18)/len(tokens_granules_17_18)
granules_17_18_pos = findCount(tokens_positive,text_granules_17_18)/len(tokens_granules_17_18)

granules_18_19_neg = findCount(tokens_negative,text_granules_18_19)/len(tokens_granules_18_19)
granules_18_19_pos = findCount(tokens_positive,text_granules_18_19)/len(tokens_granules_18_19)

granules_19_20_neg = findCount(tokens_negative,text_granules_19_20)/len(tokens_granules_19_20)
granules_19_20_pos = findCount(tokens_positive,text_granules_19_20)/len(tokens_granules_19_20)
```

# Plot

```python
year = [2018, 2019, 2020]

ajantapharma_neg = [ajantapharma_17_18_neg,ajantapharma_18_19_neg,ajantapharma_19_20_neg]
ajantapharma_pos = [ajantapharma_17_18_pos,ajantapharma_18_19_pos,ajantapharma_19_20_pos]


granules_neg = [granules_17_18_neg,granules_18_19_neg,granules_19_20_neg]
granules_pos = [granules_17_18_pos,granules_18_19_pos,granules_19_20_pos]
```
