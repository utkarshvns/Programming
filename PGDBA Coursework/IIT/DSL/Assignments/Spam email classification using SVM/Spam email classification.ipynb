```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/spambase.data',header=None)
```

```python
X = data.copy()
X = X.drop([57],axis=1)


y = data[57]
```

# Train test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn import metrics
```

```python
clf = SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred= clf.predict(X_test)
```

```python
y_pred
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("F1 Score:",metrics.f1_score(y_test, y_pred))
```

# Training Classifier

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.model_selection import GridSearchCV
```

```python
pipe = Pipeline([('scalar',StandardScaler()),
                     ('pca',PCA(n_components=57)),
                     ('classifier',SVC())])
```

# Hyperparameter tuning

```python
# Create a pipeline
pipe = Pipeline([
                     ('classifier',SVC())])
# Create dictionary with candidate learning algorithms and their hyperparameters
grid_param = [{"classifier__kernel":["linear"],
               "classifier__degree":[1],
               "classifier__C": [1,
               "classifier__gamma": ["auto"],
               "classifier__shrinking": [False]
               
              }]

# create a gridsearch of the pipeline, the fit the best model
gridsearch = GridSearchCV(pipe, grid_param, cv=2, verbose=3,n_jobs=-1) # Fit grid search
best_model = gridsearch.fit(X_train,y_train)
```

```python
gridsearch.best_estimator_
```

```python
best_model.score(X_test,y_test)
```

```python
y_pred = best_model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("F1 Score:",metrics.f1_score(y_test, y_pred))
```

# Standard scaling and PCA and then Hyperparameter tuning 

```python
pipe = Pipeline([('scalar',StandardScaler()),
                     ('pca',PCA(n_components=57)),
                     ('classifier',SVC())])

# Create dictionary with candidate learning algorithms and their hyperparameters
grid_param = [{"classifier__kernel":["linear", "poly", "rbf", "sigmoid", "precomputed"],
               "classifier__degree":[1,2,3],
               "classifier__C": [1,2,5],
               "classifier__gamma": ["scale","auto"],
               "classifier__shrinking": [True,False]
               
              }]

# create a gridsearch of the pipeline, the fit the best model
gridsearch = GridSearchCV(pipe, grid_param, cv=2, verbose=3,n_jobs=-1) # Fit grid search
best_model = gridsearch.fit(X_train,y_train)
```
