```python
# Store a variable target which takes a value 1 if the price increases with repect to prev day and 0 if it decreases
```

```python
close = dji["Close"]
```

```python
target = []
for i in range(len(close)):
    if(i==0):
        target.append(0)
    elif(close[i]>close[i-1]):
        target.append(1)
    else:
        target.append(0)    
```

```python
# Converting to date time object
# dji["Date"] = pd.to_datetime(dji["Date"] ).values.astype(float)
dji = dji.drop(['Date'],axis=1)
dji['timestamp'] = dji.index

dji = dji.drop(['Name'],axis=1)


dji['target'] = target
```
