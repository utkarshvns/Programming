# Scaled data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scale = StandardScaler()
scaled_data = scale.fit_transform(X)
scaled_data
```

```python
scaled_data = pd.DataFrame(scaled_data,columns=X.columns)
scaled_data.head()
```

```python
model = KMeans(n_clusters=3)
model.fit(scaled_data)
all_predictions = model.predict(scaled_data)
print(all_predictions)


fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

ax.scatter( scaled_data["balance"],scaled_data["age"], scaled_data["duration"],c=all_predictions)

ax.set_xlabel('balance')
ax.set_ylabel('age')
ax.set_zlabel('duration')

plt.show()
```

# PCA on data

```python
pca = PCA(n_components=3)
# pca = PCA()
pca_model = pca.fit(X)
pca_df = pca_model.transform(X)
pca_df
```

```python
pca_df = pd.DataFrame(pca_df,columns=['PC_' + str(i) for i in range(1, 4)])
pca_df.head()
```

```python
var = pca_model.explained_variance_ratio_
var
```

```python
cumulative_var = np.cumsum(np.round(var, decimals=4)*100)
plt.plot(cumulative_var,'k-o',markerfacecolor='None',markeredgecolor='k')
plt.title('Principal Component Analysis',fontsize=12)
plt.xlabel("Principal Component",fontsize=12)
plt.ylabel("Cumulative Proportion of Variance Explained",fontsize=12)
```

```python
model = KMeans(n_clusters=3)
model.fit(pca_df)
all_predictions = model.predict(pca_df)
print(all_predictions)


fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

ax.scatter( pca_df["PC_3"],pca_df["PC_2"], pca_df["PC_1"],c=all_predictions)

ax.set_xlabel('PC_3')
ax.set_ylabel('PC_2')
ax.set_zlabel('PC_1')

plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))

visualizer.fit(pca_df)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

## Deep learning

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
y = X["y_yes"]

x = X.copy()
del x["y_yes"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=25)
```

```python
test_loss, test_acc = model.evaluate(x_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```
