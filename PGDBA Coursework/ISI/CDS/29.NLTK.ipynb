## Stemming

The process of reducing inflected (or sometimes derived) words to their roots (stems). There are many approaches to stemming, but the most commonly used one is Porter's stemmer which is based on certain rules to strip suffixes. The snowball stemmer is an improved version of Poter's stemmer. 

```python
from nltk.stem import (PorterStemmer, LancasterStemmer)
from nltk.stem.snowball import SnowballStemmer 

porter = PorterStemmer()
lancaster = LancasterStemmer()
snowball = SnowballStemmer("english")

words = ['stat','statistical', 'statically', 'statistics', 'static', 'things', 'cats', 'people', 'ate', 'eat', 'gone', 'went', 'indian', 'religious',
    'be', 'am','is','defensive','defense','defence']

for w in words:
    print(w, "-->",  porter.stem(w),  snowball.stem(w), lancaster.stem(w))
```

### Note:
SnowballStemmer can stem several other languages beside English.

To make, for instance, a French stemmer, we can do the following: french_stemmer = SnowballStemmer('french')

## Lemmatizing

Lemmatization aims to achieve a similar base "stem" for a word, but aims to derive the genuine dictionary root word, not just a trunctated version of the word.  The default lemmatization method with the Python NLTK is the WordNet lemmatizer.

```python
from nltk import WordNetLemmatizer
import textblob as tb
from textblob import TextBlob, Word, Blobber

wnl = WordNetLemmatizer()

for w in words:
    print(w, "-->",  porter.stem(w), snowball.stem(w), wnl.lemmatize(w), tb.Word(w).lemmatize())
```

As we saw above, sometimes, if we try to lemmatize a word, it will end up with the same word. This is because the default part of speech is nouns.

```python
for w in words:
    print(w, "-->",  porter.stem(w), snowball.stem(w), wnl.lemmatize(w, pos='v'))
```

```python
tokens = word_tokenize(my_text)

pos_tagged_tokens = nltk.pos_tag(tokens)

print(pos_tagged_tokens)
```

```python
from nltk.corpus import wordnet

def get_pos_tag_sign(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
print(wordnet.ADV)
```

```python
for w, postag in pos_tagged_tokens:
    print(w, "(", postag, ") -->",  porter.stem(w), snowball.stem(w), wnl.lemmatize(w, pos=get_pos_tag_sign(postag)))
```
