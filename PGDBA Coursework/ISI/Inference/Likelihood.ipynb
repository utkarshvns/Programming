# Likelihood of  10 tosses with 0 heads

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# L = p^0 * (1-p)^10 = (1-p)^10
```

```python
p = np.arange(0,1,.01)
```

```python
L1 = (1-p)**10
```

```python
L1
```

```python
# plt.figure(figsize=(5, 5))
plt.plot(p,L1)
plt.show()
```

# Likelihood of  10 tosses with 5 heads

```python
# L = p^5 * (1-p)^5
```

```python
L2 = (p**5) * ((1-p)**5)
```

```python
L2
```

```python
plt.plot(p,L2)
plt.show()
```
