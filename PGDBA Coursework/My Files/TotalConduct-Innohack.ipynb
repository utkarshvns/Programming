```python
from sklearn import *
from sklearn.model_selection import *
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
import warnings
warnings.filterwarnings('ignore')
```

```python
X = null_handled_df.copy()
del X["Default_Tag"]
del X["Application ID"]
X = pd.get_dummies(X, drop_first=True)
y = np.array(null_handled_df["Default_Tag"])

del X["No of times 90 DPD or worse in last 6 months"]
del X["No of times 60 DPD or worse in last 6 months"]
del X["No of times 30 DPD or worse in last 6 months"]
del X["No of times 60 DPD or worse in last 12 months"]
del X["No of times 30 DPD or worse in last 12 months"]
del X["No of Inquiries in last 6 months (excluding home & auto loans)"]
del X["Presence of open home loan"]

X = pd.get_dummies(X, drop_first=True)
y = np.array(null_handled_df["Default_Tag"])
```

# Training Models

```python
# Logistic Regression model
```

```python
est = LogisticRegression()
pipe = Pipeline([
    ("scaler", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(degree = 1, include_bias=False)),
    ("est", est)
])

param_grid = {
    #"est__alpha": 10 ** np.linspace(-4, -2, 5)
}

grid_logistic = GridSearchCV(cv = 5, estimator=pipe, param_grid=param_grid)
grid_logistic.fit(X_train, y_train)

# y_test_pred = grid_logistic.best_estimator_.predict(X_test)

y_test_pred_prob = grid_logistic.best_estimator_.predict_proba(X_test)
threshold=0.1
y_test_pred = np.where(y_test_pred_prob[:,1] > threshold, 1, 0)

cm = confusion_matrix(y_test,y_test_pred)
print("Accuracy: ", metrics.accuracy_score(y_test, y_test_pred))
print("Recall: ", metrics.recall_score(y_test, y_test_pred))
print("Precision: ", metrics.precision_score(y_test, y_test_pred))
print("F1 score: ", metrics.f1_score(y_test, y_test_pred))
print("Confusion matrix:")
print(cm)
```

```python
# SGD classifier
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
est = SGDClassifier(loss="log", penalty="l1", max_iter=50)
pipe = Pipeline([
    ("scaler", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(degree = 1, include_bias=False)),
    ("est", est)
])

param_grid = {
    #"est__alpha": 10 ** np.linspace(-4, -2, 5)
}

grid_sgd = GridSearchCV(cv = 5, estimator=pipe, param_grid=param_grid)
grid_sgd.fit(X_train, y_train)

y_test_pred = grid_sgd.best_estimator_.predict(X_test)

y_test_pred_prob = grid_sgd.best_estimator_.predict_proba(X_test)
threshold=0.1
y_test_pred = np.where(y_test_pred_prob[:,1] > threshold, 1, 0)

cm = confusion_matrix(y_test,y_test_pred)
print("Accuracy: ", metrics.accuracy_score(y_test, y_test_pred))
print("Recall: ", metrics.recall_score(y_test, y_test_pred))
print("Precision: ", metrics.precision_score(y_test, y_test_pred))
print("F1 score: ", metrics.f1_score(y_test, y_test_pred))
print("Confusion matrix:")
print(cm)
```

```python
est = svm.SVC(gamma='scale')
pipe = Pipeline([
    ("scaler", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(degree = 1, include_bias=False)),
    ("est", est)
])

param_grid = {
    #"est__alpha": 10 ** np.linspace(-4, -2, 5)
}

grid_svm = GridSearchCV(cv = 5, estimator=pipe, param_grid=param_grid)
grid_svm.fit(X_train, y_train)

y_test_pred = grid_svm.best_estimator_.predict(X_test)

cm = confusion_matrix(y_test,y_test_pred)
print("Accuracy: ", metrics.accuracy_score(y_test, y_test_pred))
print("Recall: ", metrics.recall_score(y_test, y_test_pred))
print("Precision: ", metrics.precision_score(y_test, y_test_pred))
print("F1 score: ", metrics.f1_score(y_test, y_test_pred))
print("Confusion matrix:")
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
est = RandomForestClassifier(n_estimators=10)
pipe = Pipeline([
    ("scaler", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(degree = 1, include_bias=False)),
    ("est", est)
])

param_grid = {
    #"est__alpha": 10 ** np.linspace(-4, -2, 5)
}

grid_rf = GridSearchCV(cv = 5, estimator=pipe, param_grid=param_grid)
grid_rf.fit(X_train, y_train)

y_test_pred = grid_rf.best_estimator_.predict(X_test)

y_test_pred_prob = grid_rf.best_estimator_.predict_proba(X_test)
threshold=0.1
y_test_pred = np.where(y_test_pred_prob[:,1] > threshold, 1, 0)


cm = confusion_matrix(y_test,y_test_pred)
print("Accuracy: ", metrics.accuracy_score(y_test, y_test_pred))
print("Recall: ", metrics.recall_score(y_test, y_test_pred))
print("Precision: ", metrics.precision_score(y_test, y_test_pred))
print("F1 score: ", metrics.f1_score(y_test, y_test_pred))
print("Confusion matrix:")
print(cm)
```
